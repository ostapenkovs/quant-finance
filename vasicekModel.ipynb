```python
import numpy as np
from numpy.random import normal
import matplotlib.pyplot as plt
```

```python
def generate_process(dt=0.1, theta=1.2, mu=0.5, sigma=0.1, n=10000):
    x = np.zeros(n)

    for t in range(1, n):
        x[t] = x[t-1] + theta*(mu-x[t-1])*dt + sigma * normal(0, np.sqrt(dt))
    
    return x
```

```python
def plot_process(x):
    plt.figure(figsize=(12, 5))
    plt.plot(x)
    plt.xlabel("t")
    plt.ylabel("x(t)")
    plt.title("Ornstein-Uhlenbeck Process")
    plt.show()
```

```python
def main():
    data = generate_process()
    plot_process(data)
```

```python
if __name__ == "__main__":
    main()
```
