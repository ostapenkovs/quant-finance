```python
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
```

```python
NUM_SIMULATIONS = 10000
NUM_TOTAL = 1000

x = np.zeros(NUM_TOTAL)
y = np.zeros(NUM_TOTAL)
pik = np.zeros(NUM_SIMULATIONS)
rollavg = np.zeros(NUM_SIMULATIONS)

for i in range(NUM_SIMULATIONS):
    num_within = 0
    for j in range(NUM_TOTAL):
        x[j], y[j] = np.random.rand(), np.random.rand()
        if y[j] < sqrt(1-x[j]**2):
            num_within += 1
    pik[i] = num_within / NUM_TOTAL * 4
    rollavg[i] = pik[:i+1].mean()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(rollavg)
```
